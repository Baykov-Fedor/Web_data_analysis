Найдем средний возраст (год) топ лучших фильмов с Кинопоиска

```python
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup

top = urlopen('https://www.kinopoisk.ru/top/') # скачиваем страничку со списком топ-250 фильмов по версии кинопоиска
html = top.read().decode('utf8') # считываем содержимое
soup = BeautifulSoup(html, 'html.parser') # делаем суп
```

```python
print(soup.prettify())
```

```python
import re
import statistics
from collections import Counter

years = []
for films in soup.find_all('a', class_='all'):
    name = films.get_text()
    result = re.search(r'\(\d{4}\)$', name) #извлекаем из конца названия фильма "(год)"
    if result:
        temp_str = result.group()
    
        result = re.search(r'\d{4}', temp_str) #извлекаем только цифры чтобы сделать int
        matched_result = result.group()
        years.append(int(matched_result)) #полученный год производства добавляем в список

print(f'В списке: {len(years)} фильмов')
print(f'Медианная дата производства фильмов: {statistics.median(years)}') 
print(f'Средняя дата производства фильмов: {sum(years)/len(years)}') 
print(f'Больше всего "шедевров" вышло в: {Counter(years).most_common()[0]} и в {Counter(years).most_common()[1]}')
```
